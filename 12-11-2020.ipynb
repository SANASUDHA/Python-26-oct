### Data Analysis and Manipulation

```python
df1 = pd.DataFrame({'no_persons':[12,13,14,15],
                   'no_days':[22,33,45,67]},
                  index = [2019,2020,2007,2010])
df1
```

```python
df2 = pd.DataFrame({'mve_rating':[3.4,6.5,4.6,4.4],
                   'no_heros':[1,3,2,5]},
                  index = [2019,2020,2030,2013])
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df3 = pd.DataFrame({'a':[80,90,60,70],
                   'b':[2,1,2,3],
                   'c':[50,45,43,67]},
                  index = [2001,2002,2003,2004])
df3
```

```python
df4 = pd.DataFrame({'a':[80,69,60,70],
                   'b':[2,1,2,3],
                   'c':[50,45,43,67]},
                  index = [2001,2002,2003,2004])
df4
```

```python
df3.merge(df4)
```

```python
con = pd.concat([df1,df2])
con
```
